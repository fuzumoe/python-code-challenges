````python
"""
Solve the problem [Python]
Given a binary tree of integers root, create 3 functions to print the tree nodes in preorder, inorder, and postorder traversal.
Preorder: print the root value, then print the left subtree, then print the right subtree.
Inorder: print the left subtree, then print the root value, then print the right subtree.
Postorder: print the left subtree, then print the right subtree, then print the root value.

Example 1:

Input: root = [1, 2, 3, 4, 5, 6, 7]

Output:
Preorder: 1 2 4 5 3 6 7
Inorder: 4 2 5 1 6 3 7
Postorder: 4 5 2 6 7 3 1
"""

class Tree:
    def __init__(self, data, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right

def dfsPreorder(root):
    if root:
        print(root.data, end=' ')
        dfsPreorder(root.left)
        dfsPreorder(root.right)

def dfsInorder(root):
    if root:
        dfsInorder(root.left)
        print(root.data, end=' ')
        dfsInorder(root.right)

def dfsPostorder(root):
    if root:
        dfsPostorder(root.left)
        dfsPostorder(root.right)
        print(root.data, end=' ')

# Example usage:
# Construct the tree: [1, 2, 3, 4, 5, 6, 7]
root = Tree(1, Tree(2, Tree(4), Tree(5)), Tree(3, Tree(6), Tree(7)))

print("Preorder: ", end='')
dfsPreorder(root)
print("\nInorder: ", end='')
dfsInorder(root)
print("\nPostorder: ", end='')
dfsPostorder(root)
````

````python
""" 
Solve the problem [Python]
Given a non-empty array of integers arr, create a function that returns the sum of the subarray that has the greatest sum.
We don't consider the empty array [] as a subarray.

Example 1:

Input: arr = [2, 3, -6, 4, 2, -8, 3]

Output: 6

Explanation: the maximum subarray is [4, 2], its sum is 6

Example 2:

Input: arr = [2, 3, -1, 4, -10, 2, 5]

Output: 8

Explanation: the maximum subarray is [2, 3, -1, 4], its sum is 8

Example 3:

Input: arr = [-3, -1, -2]

Output: -1

Explanation: the maximum subarray is [-1], its sum is -1
"""
def max_subarray_sum(arr):
    max_sum = arr[0]
    current_sum = arr[0]
    
    for num in arr[1:]:
        current_sum = max(num, current_sum + num)
        max_sum = max(max_sum, current_sum)
    
    return max_sum

# Example usage:
display(max_subarray_sum([2, 3, -6, 4, 2, -8, 3]))  # Output: 6
display(max_subarray_sum([2, 3, -1, 4, -10, 2, 5]))  # Output: 8
display(max_subarray_sum([-3, -1, -2]))  # Output: -1
````

````python
"""
Solve the problem [Python]
Given a binary tree of integers root, create a function that reverses it left to right in-place.

Example 1:

Input: root = [1, 2, 3, 4, 5, 6, 7]

Output: [1, 3, 2, 7, 6, 5, 4]

Example 2:

Input: root = [6, 8, 13, 2, 1, 5, null, 7]

Output: [6, 13, 8, null, 5, 1, 2, null, null, null, null, null, 7]

"""

class Tree:
    def __init__(self, data, left = None, right = None):
        self.data = data
        self.left = left
        self.right = right

# Parameters:
#  root: Tree
# return type: void

def reverseTree(root):
    if root is None:
        return
    
    # Swap left and right children
    root.left, root.right = root.right, root.left

    # Recursively call the function on the children
    reverseTree(root.left)
    reverseTree(root.right)

def printTree(root):
    if root is None:
        return
    # Print the current node's data
    print(root.data, end=" ")
    # Recursively print the left and right children
    printTree(root.left)
    printTree(root.right)
    
# Example usage:
# Creating the tree from example 1
root = Tree(1, Tree(2, Tree(4), Tree(5)), Tree(3, Tree(6), Tree(7)))
reverseTree(root)
display(printTree(root))  # Output: 1 3 7 6 2 5 4

# Creating the tree from example 2
root = Tree(6, Tree(8, Tree(2, Tree(7)), Tree(1)), Tree(13, Tree(5)))
reverseTree(root)
display(printTree(root))  # Output: 6 13 5 8 1 2 7
````

````python
"""
Solve the problem [Python]
Given a string str made of alphabetical letters only, create a function that returns the length of the longest substring without repeating characters.

Example 1:

Input: str = "abcdbeghef"

Output: 6

Explanation: the longest substring without repeating characters is "cdbegh", its length is 6

Example 2:

Input: str = "aaaaa"

Output: 1

Explanation: the longest substring without repeating characters is "a", its length is 1

Example 3:

Input: str = "eddy"

Output: 2

Explanation: the longest substring without repeating characters is "ed" (or "dy"), its length is 2

"""
def longestSubstringWithoutRepeating(s):
    max_length = 0
    start = 0
    char_index_map = {}
    
    for i, char in enumerate(s):
        if char in char_index_map and char_index_map[char] >= start:
            start = char_index_map[char] + 1
        char_index_map[char] = i
        max_length = max(max_length, i - start + 1)
    
    return max_length

# Example usage:
display(longestSubstringWithoutRepeating("abcdbeghef"))  # Output: 6
display(longestSubstringWithoutRepeating("aaaaa"))  # Output: 1
display(longestSubstringWithoutRepeating("eddy"))  # Output: 2
````

````python
""" 
Solve the problem [Python]
Given a linked list of integers list, create a function that reverses it in-place without using an additional data structure.

Example 1:

    Input: list = 5 -> 3 -> 6 -> 4 -> 7 -> null

    Output: 7 -> 4 -> 6 -> 3 -> 5 -> null

Example 2:

    Input: list = 1 -> 2 -> 3 -> null

    Output: 3 -> 2 -> 1 -> null
"""
class Node:
    def __init__(self, data, next=None):
        self.data = data
        self.next = next

class LinkedList:
    def __init__(self, head=None):
        self.head = head

def reverseList(list):
    # Initialize prev to None, which will eventually become the new tail of the list
    prev = None
    # Initialize current to the head of the list
    current = list.head
    
    # Iterate through the list until current is None (end of the list)
    while current is not None:
        # Store the next node in a temporary variable
        next_node = current.next
        # Reverse the link by pointing the current node's next to the previous node
        current.next = prev
        # Move prev and current one step forward
        prev = current
        current = next_node
        
    # After the loop, prev will be the new head of the reversed list
    list.head = prev

# Example usage:
def printList(head):
    current = head
    while current:
        print(current.data, end=" -> ")
        current = current.next
    print("null")

# Creating the linked list 5 -> 3 -> 6 -> 4 -> 7 -> null
node5 = Node(7)
node4 = Node(4, node5)
node3 = Node(6, node4)
node2 = Node(3, node3)
node1 = Node(5, node2)
linked_list = LinkedList(node1)

print("Original list:")
printList(linked_list.head)

reverseList(linked_list)

print("Reversed list:")
printList(linked_list.head)
````

````python
""" 
Solve the problem [Python]
Given a non-empty array of non-negative integers arr, 
where each element represents the maximum jump that we can perform from that index, 
create a boolean function that checks if we can reach the last index starting from the first one.

Example 1:

Input: arr = [3, 0, 2, 0, 2, 1, 4, 3]

Output: true

Explanation: we can for example jump from arr[0] to arr[2], then from arr[2] to arr[4], then from arr[4] to arr[6], then from arr[6] to arr[7] (the last index)
"""
def canJump(arr):
    max_reachable = 0
    n = len(arr)
    
    for i in range(n):
        if i > max_reachable:
            return False
        max_reachable = max(max_reachable, i + arr[i])
    
    return True
"""
### Example

**Input**: `arr = [3, 0, 2, 0, 2, 1, 4, 3]`

**Output**: `True`

**Explanation**: We can jump from `arr[0]` to `arr[2]`, then from `arr[2]` to `arr[4]`, then from `arr[4]` to `arr[6]`, and finally from `arr[6]` to `arr[7]` (the last index).

### Solution Explanation

To solve this problem, we use a greedy algorithm. The idea is to keep track of the farthest index we can reach as we iterate through the array. If at any point the current index is greater than the farthest index we can reach, it means we cannot proceed further and hence cannot reach the last index.

#### Steps

1. **Initialize**:
   - Create a variable `max_reachable` and set it to 0. This variable will keep track of the farthest index we can reach.

2. **Iterate through the array**:
   - For each index `i` in the array:
     - Check if `i` is greater than `max_reachable`. If it is, return `False` because we cannot proceed further.
     - Update `max_reachable` to be the maximum of `max_reachable` and `i + arr[i]`.

3. **Completion**:
   - If we complete the loop, it means we can reach the last index, so return `True`.

### Code Implementation

```python
def canJump(arr):
    max_reachable = 0
    n = len(arr)
    
    for i in range(n):
        if i > max_reachable:
            return False
        max_reachable = max(max_reachable, i + arr[i])
    
    return True

# Example usage
arr = [3, 0, 2, 0, 2, 1, 4, 3]
print("Input:", arr)
print("Output:", canJump(arr))  # Output: True
```

### Detailed Explanation

- **Initialization**:
  - `max_reachable` is initialized to 0. This means initially we can only reach the first index.

- **Iteration**:
  - We iterate through the array using a for loop.
  - For each index `i`:
    - If `i` is greater than `max_reachable`, it means we cannot reach this index from any of the previous indices, so we return `False`.
    - Otherwise, we update `max_reachable` to be the maximum of `max_reachable` and `i + arr[i]`. This step ensures that we are always aware of the farthest index we can reach from the current index.

- **Completion**:
  - If we complete the loop without returning `False`, it means we can reach the last index, so we return `True`.

### Example Walkthrough

For the input `arr = [3, 0, 2, 0, 2, 1, 4, 3]`:

1. **Initialization**:
   - `max_reachable = 0`

2. **Iteration**:
   - `i = 0`: `max_reachable = max(0, 0 + 3) = 3`
   - `i = 1`: `max_reachable = max(3, 1 + 0) = 3`
   - `i = 2`: `max_reachable = max(3, 2 + 2) = 4`
   - `i = 3`: `max_reachable = max(4, 3 + 0) = 4`
   - `i = 4`: `max_reachable = max(4, 4 + 2) = 6`
   - `i = 5`: `max_reachable = max(6, 5 + 1) = 6`
   - `i = 6`: `max_reachable = max(6, 6 + 4) = 10`
   - `i = 7`: `max_reachable = max(10, 7 + 3) = 10`

3. **Completion**:
   - We complete the loop without returning `False`, so we return `True`.

This approach ensures that we can determine if we can reach the last index efficiently with a time complexity of O(n).
"""
# Example usage
arr = [3, 0, 2, 0, 2, 1, 4, 3]
print("Input:", arr)
print("Output:", canJump(arr))  # Output: True
````
